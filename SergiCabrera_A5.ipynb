```python
class Song:
    def __init__(self, titol, autor, durada):
        self.titol = titol
        self.autor = autor
        self.durada = durada  

    def info(self):
        return f"{self.titol} - {self.autor} ({self.durada} segons)"

class Playlist:
    def __init__(self):
        self.song_list = []

    def afegir_canco(self, song):
        self.song_list.append(song)

    def durada_total(self):
        total = sum(s.durada for s in self.song_list)
        return total

    def mostrar_playlist(self):
        for s in self.song_list:
            print(s.info())

    def canco_mes_llarga(self):
        if not self.song_list:
            return None
   
        mes_llarga = max(self.song_list, key=lambda s: s.durada)
        return mes_llarga.info()

c1 = Song("Diamantito", "Mikecrack", 354)
c2 = Song("67", "67 Guy", 183)

la_meva_llista = Playlist()
la_meva_llista.afegir_canco(c1)
la_meva_llista.afegir_canco(c2)

c3 = Song("Merluzito", "Roger COLLAB Motoner", 157)
la_meva_llista.afegir_canco(c3)

print("    Totes les cançons    ")
la_meva_llista.mostrar_playlist()

print(f"\nDurada total: {la_meva_llista.durada_total()} segons")
print(f"Cançó més llarga: {la_meva_llista.canco_mes_llarga()}") 
```

```python
class Personatge:
    def __init__(self, nom, vida, forca, nivell):
        self.nom = nom
        self.vida = vida
        self.forca = forca
        self.nivell = nivell

    def atacar(self, enemic):
        print(f"{self.nom} ataca a {enemic.nom}!")
        enemic.vida -= self.forca
        if enemic.vida < 0:
            enemic.vida = 0
        print(f"Vida restant de {enemic.nom}: {enemic.vida}")

    def pujar_nivell(self):
        self.nivell += 1
        self.forca += 2
        self.vida += 5
        print(f"{self.nom} ha pujat al nivell {self.nivell}!")

    def esta_viu(self):
        return self.vida > 0

    def __str__(self):
        return f"PERSONATGE: {self.nom} | Vida: {self.vida} | Força: {self.forca} | Nivell: {self.nivell}"

p1 = Personatge("Messi", 30, 10, 1)
p2 = Personatge("Ronaldo", 25, 8, 1)

torn = 1
while p1.esta_viu() and p2.esta_viu():
    print(f"\n   Torn {torn}    ")
  
    p1.atacar(p2)
   
    if p2.esta_viu():
        p2.atacar(p1)
   
    torn += 1

if p1.esta_viu():
    guanyador = p1
else:
    guanyador = p2

print(f"\nEl guanyador és {guanyador.nom}!")
guanyador.pujar_nivell()

print(guanyador)
```
